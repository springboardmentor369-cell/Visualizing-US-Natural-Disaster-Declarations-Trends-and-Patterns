```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("raw data.txt")
```

```python
df.head()
df.shape
```

```python
print(df.head())
```

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
```

```python
df.drop(columns=['id', 'incidentid'], inplace=True, errors='ignore')
```

```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
cat_cols = ['state', 'stateabbreviation', 'incidentname',
            'requestedincidenttypes', 'declarationrequesttype',
            'currentrequeststatus']

for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].astype(str).str.strip()
```

```python
date_cols = [
    'declarationrequestdate',
    'requestedincidentbegindate',
    'requestedincidentenddate',
    'incidentbegindate',
    'requeststatusdate'
]

for col in date_cols:
    if col in df.columns:
        df[col] = pd.to_datetime(df[col], errors='coerce')
```

```python
df.isnull().sum()
```

```python
df.fillna({
    'requestedincidentenddate': pd.NaT,
    'incidentname': 'Unknown',
    'requestedincidenttypes': 'Unknown'
}, inplace=True)
```

```python
program_cols = [
    'ihprogramrequested',
    'iaprogramrequested',
    'paprogramrequested',
    'hmprogramrequested'
]

for col in program_cols:
    if col in df.columns:
        df[col] = df[col].fillna(0).astype(int)
```

```python
df['requestedincidenttypes'] = (
    df['requestedincidenttypes']
    .str.lower()
    .str.replace('&', 'and')
    .str.title()
)
```

```python
df['year'] = df['declarationrequestdate'].dt.year
df['month'] = df['declarationrequestdate'].dt.month
```

```python
df.to_csv("/content/cleaned_disaster_data.csv", index=False)
```
